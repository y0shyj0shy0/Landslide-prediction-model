```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# 훈련 데이터를 이용하여 MinMaxScaler를 초기화하고 훈련
scaler = MinMaxScaler()
train_data = pd.read_parquet("/home/kng/kng/MOUNTAIN2/trainDatas/SMOTE/REALS_80.parquet")  # 훈련 데이터 읽기
normalized_train_data = scaler.fit_transform(train_data)  # 훈련 데이터로 스케일러 훈련

# 테스트 데이터를 정규화
test_data = pd.read_parquet("/home/kng/kng/MOUNTAIN2/testDatas/REALS_20.parquet")  # 테스트 데이터 읽기
normalized_test_data = scaler.transform(test_data)  # 테스트 데이터 정규화

# 정규화된 테스트 데이터를 데이터프레임으로 변환
normalized_train_df = pd.DataFrame(normalized_train_data, columns=train_data.columns)
normalized_test_df = pd.DataFrame(normalized_test_data, columns=test_data.columns)

# 정규화된 테스트 데이터 출력
#print("Normalized test data:\n", normalized_test_df)

normalized_train_df.to_parquet('/home/kng/kng/MOUNTAIN2/trainDatas/SMOTE/normalized/NORM_REALS_80.parquet',index=False)
normalized_test_df.to_parquet('/home/kng/kng/MOUNTAIN2/testDatas/SMOTE/normalized/NORM_REALS_20.parquet', index=False)

```

```python
import pandas as pd

# Parquet 파일 불러오기
test_data = pd.read_parquet('/home/kng/kng/MOUNTAIN2/testDatas/SMOTE/final/SMOTED_FINAL_20.parquet')

# is_landslides 열의 값이 1인 행 추출
landslides_data = test_data[test_data['is_landslided'] == 1]

# 1인 값의 개수
num_landslides = landslides_data.shape[0]

# is_landslides 열의 값이 0인 행 추출
non_landslides_data = test_data[test_data['is_landslided'] == 0]

# 1의 개수에 맞게 0인 행 삭제
if num_landslides < non_landslides_data.shape[0]:
    # 1의 개수에 맞게 0인 행을 삭제
    non_landslides_data = non_landslides_data.sample(n=num_landslides, random_state=42)

# 수정된 데이터프레임 생성
balanced_data = pd.concat([landslides_data, non_landslides_data], ignore_index=True)

# 새로운 Parquet 파일로 저장
balanced_data.to_parquet('/home/kng/kng/MOUNTAIN2/testDatas/SMOTE/final/balanced_data.parquet', index=False)
```
